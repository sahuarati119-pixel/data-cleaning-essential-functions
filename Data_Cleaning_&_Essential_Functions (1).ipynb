3.Explain the use of text functions such as TRIM, LEFT, RIGHT, MID, and CONCAT in data cleaning.

*   TRIM: Removes extra spaces from text except single spaces between words. Ex: =TRIM("  Hello  ") → Hello
*   LEFT: Extracts characters from the start. Ex: =LEFT("Excel",2) → Ex
*   RIGHT: Extracts characters from the end. Ex: =RIGHT("Excel",2) → el
*   MID: Extracts from the middle. Ex: =MID("Excel",2,3) → xce
*   CONCAT: Joins multiple text strings. Ex: =CONCAT("Hello"," ","World") → Hello World
